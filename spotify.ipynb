### Dont run this cell unless you want to try and pull all posible lyrics for the df2, this will return an np.nan for any nonetype object

```python
#Saving all lyrics from dataset to a list, was only able to grab ~600 before getting cut off

lyrics_ls = []

for x, y in art_track.itertuples(index=False):
    song = genius.search_song(y, x)
    if song == None:
        lyrics_ls.append(np.nan)
    else:
        lyrics_ls.append(song.lyrics)
```

```python
# Create a series to save the small amount of results
samp = pd.Series(lyrics_ls)
```

```python
#Since lyric_ls length is only 659 I chop down df2 so I dont run into dimmension issues
lyric_df = df2[:659]
lyric_df["lyric"] = lyrics_ls
lyric_df.to_csv("lyrics_df.csv")
```
